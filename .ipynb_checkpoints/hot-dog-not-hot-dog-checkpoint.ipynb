```python
import os
import shutil
from tqdm import tqdm
from random import shuffle
import cv2
import numpy as np

TRAIN_HOT_DOG_DIR = "hot-dog-not-hot-dog/train/hot_dog"
TRAIN_NOT_HOT_DOG_DIR = "hot-dog-not-hot-dog/train/not_hot_dog"

TEST_HOT_DOG_DIR = "hot-dog-not-hot-dog/test/hot_dog"
TEST_NOT_HOT_DOG_DIR = "hot-dog-not-hot-dog/test/not_hot_dog"

IMAGE_SIZE = 72
LR = 1e-5
MODEL_NAME = "hot-dog-not-hot-dog_{}".format("version-1")
```

```python
def transfer_data():
    count=0
    for img in tqdm(os.listdir(TEST_HOT_DOG_DIR)):
        if count<200:
            shutil.move(os.path.join(TEST_HOT_DOG_DIR,img),TRAIN_HOT_DOG_DIR)
            count=count+1
    count=0
    for img in tqdm(os.listdir(TEST_NOT_HOT_DOG_DIR)):
        if count<200:
            shutil.move(os.path.join(TEST_NOT_HOT_DOG_DIR,img),TRAIN_NOT_HOT_DOG_DIR)
            count=count+1

transfer_data()
```

```python
print(len(os.listdir(TRAIN_HOT_DOG_DIR)))
print(len(os.listdir(TRAIN_NOT_HOT_DOG_DIR)))
print(len(os.listdir(TEST_HOT_DOG_DIR)))
print(len(os.listdir(TEST_NOT_HOT_DOG_DIR)))
```

```python
def create_train_data():
    
    #hot-dog = [1,0]
    #not-hot-dog = [0,1]
    
    training_data = []
    for img in tqdm(os.listdir(TRAIN_HOT_DOG_DIR)):
        label = [1,0]
        path = os.path.join(TRAIN_HOT_DOG_DIR, img)
        img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
        img = cv2.resize(img, (IMAGE_SIZE, IMAGE_SIZE))
        training_data.append([np.array(img), np.array(label)])
    
    for img in tqdm(os.listdir(TRAIN_NOT_HOT_DOG_DIR)):
        label = [0,1]
        path = os.path.join(TRAIN_NOT_HOT_DOG_DIR, img)
        img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
        img = cv2.resize(img, (IMAGE_SIZE, IMAGE_SIZE))
        training_data.append([np.array(img), np.array(label)])
    
    shuffle(training_data)
    return training_data
```

```python
def create_validation_and_test_data():
    
    #hot-dog = [1,0]
    #not-hot-dog = [0,1]
    count=0
    testing_data = []
    validation_data = []
    for img in tqdm(os.listdir(TEST_HOT_DOG_DIR)):
        if count<25:
            label = [1,0]
            path = os.path.join(TEST_HOT_DOG_DIR, img)
            img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
            img = cv2.resize(img, (IMAGE_SIZE, IMAGE_SIZE))
            validation_data.append([np.array(img), np.array(label)])
            count=count+1
        else:
            label = [1,0]
            path = os.path.join(TEST_HOT_DOG_DIR, img)
            img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
            img = cv2.resize(img, (IMAGE_SIZE, IMAGE_SIZE))
            testing_data.append([np.array(img), np.array(label)])
            count=count+1
            
    count=0
    for img in tqdm(os.listdir(TEST_NOT_HOT_DOG_DIR)):
        if count<25:
            label = [0,1]
            path = os.path.join(TEST_NOT_HOT_DOG_DIR, img)
            img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
            img = cv2.resize(img, (IMAGE_SIZE, IMAGE_SIZE))
            validation_data.append([np.array(img), np.array(label)])
            count=count+1
        else:
            label = [0,1]
            path = os.path.join(TEST_NOT_HOT_DOG_DIR, img)
            img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
            img = cv2.resize(img, (IMAGE_SIZE, IMAGE_SIZE))
            testing_data.append([np.array(img), np.array(label)])
            count=count+1
    
    shuffle(testing_data)
    shuffle(validation_data)
    return [validation_data, testing_data]
```

```python
train_data = create_train_data()
np.save("train_data.npy",train_data)
```

```python
data = create_validation_and_test_data()
validation_data = data[0]
test_data = data[1]
np.save("validation_data.npy",data[0])
np.save("test_data.npy",data[1])
```

```python
import tflearn
from tflearn.layers.conv import conv_2d, max_pool_2d
from tflearn.layers.core import input_data, dropout, fully_connected
from tflearn.layers.estimator import regression

import tensorflow as tf
tf.reset_default_graph()

convnet = input_data(shape=[None, IMAGE_SIZE, IMAGE_SIZE, 1], name="input")

convnet = conv_2d(convnet, nb_filter=32, filter_size=5, activation="relu", strides=1, padding="same")
convnet = max_pool_2d(convnet, 3)

convnet = conv_2d(convnet, nb_filter=64, filter_size=5, activation="relu", strides=1, padding="same")
convnet = max_pool_2d(convnet, 3)

convnet = conv_2d(convnet, nb_filter=128, filter_size=5, activation="relu", strides=1, padding="same")
convnet = max_pool_2d(convnet, 3)

convnet = conv_2d(convnet, nb_filter=64, filter_size=5, activation="relu", strides=1, padding="same")
convnet = max_pool_2d(convnet, 3)

convnet = fully_connected(convnet, 1024, activation="relu")
convnet = dropout(convnet, 0.5)

convnet = fully_connected(convnet, 2, activation="softmax")
convnet = regression(convnet, optimizer="rmsprop", learning_rate=LR, loss="categorical_crossentropy", name="targets")

model = tflearn.DNN(convnet, tensorboard_dir="log")
```

```python
X = np.array([i[0] for i in train_data]).reshape(-1, IMAGE_SIZE, IMAGE_SIZE, 1)
Y = np.array([i[1] for i in train_data])

validation_x = np.array([i[0] for i in validation_data]).reshape(-1, IMAGE_SIZE, IMAGE_SIZE, 1)
validation_y = np.array([i[1] for i in validation_data])


print(X.shape)
print(Y.shape)
print(validation_x.shape)
print(validation_y.shape)
```

```python
model.fit({"input": X}, 
          {"targets": Y},
          n_epoch=5,
          validation_set=({"input": validation_x}, {"targets": validation_y}),
          show_metric=True, run_id=MODEL_NAME)
```

```python
model.save(MODEL_NAME)
```

# Using MobileNet

```python
from __future__ import absolute_import, division, print_function
import os
import tensorflow as tf
import numpy
import matplotlib.pyplot as plt

TRAIN_HOT_DOG_DIR_MOBILE_NET = "hot-dog-not-hot-dog-mobile-net/train_mobile_net/hot_dog"
TRAIN_NOT_HOT_DOG_DIR_MOBILE_NET = "hot-dog-not-hot-dog-mobile-net/train_mobile_net/not_hot_dog"

TEST_HOT_DOG_DIR_MOBILE_NET = "hot-dog-not-hot-dog-mobile-net/test_mobile_net/hot_dog"
TEST_NOT_HOT_DOG_DIR_MOBILE_NET = "hot-dog-not-hot-dog-mobile-net/test_mobile_net/not_hot_dog"

TRAIN_DIR_MOBILE_NET = "hot-dog-not-hot-dog-mobile-net/train_mobile_net"
VALIDATION_DIR_MOBILE_NET = "hot-dog-not-hot-dog-mobile-net/test_mobile_net"
```

```python
def create_train_validation_data_mobile_net():
    
    for img in tqdm(os.listdir(TRAIN_HOT_DOG_DIR)[-25:]):
        shutil.copy(os.path.join(TRAIN_HOT_DOG_DIR, img),TRAIN_HOT_DOG_DIR_MOBILE_NET)
            
    for img in tqdm(os.listdir(TRAIN_NOT_HOT_DOG_DIR)[-25:]):
        shutil.copy(os.path.join(TRAIN_NOT_HOT_DOG_DIR, img),TRAIN_NOT_HOT_DOG_DIR_MOBILE_NET)
            
    for img in tqdm(os.listdir(TEST_HOT_DOG_DIR)[-10:]):
        shutil.copy(os.path.join(TEST_HOT_DOG_DIR, img),TEST_HOT_DOG_DIR_MOBILE_NET)
    
    for img in tqdm(os.listdir(TEST_NOT_HOT_DOG_DIR)[-10:]):
        shutil.copy(os.path.join(TEST_NOT_HOT_DOG_DIR, img),TEST_NOT_HOT_DOG_DIR_MOBILE_NET)

create_train_validation_data_mobile_net()
```

```python
print("Training Hot Dog images : ",len(os.listdir(TRAIN_HOT_DOG_DIR_MOBILE_NET)))
print("Training Not Hot Dog images : ",len(os.listdir(TRAIN_NOT_HOT_DOG_DIR_MOBILE_NET)))
print("Validation Hot Dog images : ",len(os.listdir(TEST_HOT_DOG_DIR_MOBILE_NET)))
print("Validation Not Hot Dog images : ",len(os.listdir(TEST_NOT_HOT_DOG_DIR_MOBILE_NET)))
```

```python
IMAGE_SIZE_NEW = 224
BATCH_SIZE = 10

train_datagen = keras.preprocessing.image.ImageDataGenerator(preprocessing_function=keras.applications.mobilenet.preprocess_input)
validation_datagen = keras.preprocessing.image.ImageDataGenerator(preprocessing_function=keras.applications.mobilenet.preprocess_input)

train_generator = train_datagen.flow_from_directory(
    TRAIN_DIR_MOBILE_NET,
    target_size = (IMAGE_SIZE_NEW, IMAGE_SIZE_NEW),
    batch_size = BATCH_SIZE,
)
validation_generator = validation_datagen.flow_from_directory(
    VALIDATION_DIR_MOBILE_NET,
    target_size = (IMAGE_SIZE_NEW, IMAGE_SIZE_NEW),
    batch_size = BATCH_SIZE,
)
```

```python
import keras

mobile = keras.applications.mobilenet.MobileNet()
mobile.summary()
```

```python
from keras.models import Model
from keras.layers import Input, Dense

x = mobile.layers[-6].output
predictions = Dense(2, activation="softmax")(x)
model = Model(inputs=mobile.input, outputs=predictions)
model.summary()
```

```python
for layer in model.layers[:-5]:
    layer.trainable = False
```

```python
model.compile(keras.optimizers.Adam(lr=0.0001), loss="categorical_crossentropy", metrics=["accuracy"])
```

```python
epochs = 5
steps_per_epoch = train_generator.n 
validation_steps = validation_generator.n

history = model.fit_generator(train_generator,
                              steps_per_epoch = steps_per_epoch,
                              epochs = epochs,
                              validation_data = validation_generator,
                              validation_steps = validation_steps)
```

```python
model_json = model.to_json()
with open("{}.json".format(MODEL_NAME), "w") as json_file:
    json_file.write(model_json)
model.save_weights("{}.h5".format(MODEL_NAME))
```

```python
print("Training Accuracy: ",history.history['acc'][4])
print("Training Loss: ",history.history['loss'][4])
print("Validation Accuracy ",history.history['val_acc'][4])
print("Validation Loss: ",history.history['val_acc'][4])
```

```python
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
from keras.preprocessing import image
from IPython.display import Image

def predict_image(img_path):
    img = image.load_img(img_path, target_size=(IMAGE_SIZE_NEW, IMAGE_SIZE_NEW))
    img_data = image.img_to_array(img)
    img_data = np.expand_dims(img_data, axis=0)         
    img_data /= 255.
    pred = model.predict(img_data)
    plt.imshow(img_data[0])
    if np.argmax(pred[0]) == 0:
        predicted="Hot Dog, Confidence: "+str(pred[0][np.argmax(pred[0])])
    else:
        predicted="Not Hot Dog, Confidence: "+str(pred[0][np.argmax(pred[0])])
    plt.title(predicted)
    plt.axis('off')
    plt.show()
```
